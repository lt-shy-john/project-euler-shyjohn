```python
from math import sqrt
from itertools import count, islice
```

```python
number = 1
prime_number = 0

def isPrime(n):
    return n > 1 and all(n%i for i in islice(count(2), int(sqrt(n)-1)))

while number < 1000:
    print ('Run ' + repr(number))
    if isPrime(number):
        prime_number = prime_number + number
    print ('Sum of prime number is ' + repr(prime_number))
    number += 1
```
